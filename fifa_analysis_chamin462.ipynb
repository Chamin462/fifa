# Fifa Football World Cup Analysis

```python
# import libraries / loading the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the dataset / loading the dataset
fifa = pd.read_csv("players_20.csv")
```

```python
 # print the first 5 rows, Check / Printing first / top five rows, of dataset fifa
 fifa.head()
```

```python
# check the column names of the dataset
for col in fifa.columns:
    print(col)
```

```python
# number of rows and columns you know number of columns and number of rows of dataset
fifa.shape
```

```python
# check the dataset i know how country more player belong you counts 
fifa['nationality'].value_counts()
```

```python
# top ten country belong more player check
fifa['nationality'].value_counts()[0:10]
```

```python
# check top five country belong more player
fifa['nationality'].value_counts()[0:5]
```

```python
# when made bar plot then keys and value individual extract and give country name 
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# remember when make bar plot first value/parameter categorical and second value/parameter numerical / visualize check
plt.figure(figsize=(10,6))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='g')
plt.show()
```

```python
# when check how player most salary in fifa 
player_salary = fifa[['short_name','wage_eur']]
```

```python
 # check first five record of player salary
 player_salary.head()
```

```python
# sort salary how player more salary
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
# check top five player more salary in fifa
player_salary.head()
```

```python
# visualize check how player more salary / remember when make bar plot first value/parameter categorical and second value/parameter numerical 
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=['green','blue','red','orange','pink'])
plt.show()
```
